```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures,OneHotEncoder
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report,precision_score,recall_score,f1_score
```

```python
df=pd.read_csv('campus_placement_data.csv')
```

```python
x=df.drop(['salary_lpa','student_id','placed'],axis=1)
y=df['placed']          
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
cat_col=x.select_dtypes(include='object').columns
cat_col
```

```python
encoder=OneHotEncoder()
model=LogisticRegression()
```

```python
t_encoder=encoder.fit_transform(x_train[cat_col])
tt_encoder=encoder.fit_transform(x_test[cat_col])
```

```python
t_cat_name=encoder.get_feature_names_out()
tt_cat_name=encoder.get_feature_names_out()
t_value=t_encoder.toarray()
tt_value=tt_encoder.toarray()
```

```python
t_cat_df=pd.DataFrame(t_value,columns=t_cat_name)
```

```python
tt_cat_df=pd.DataFrame(tt_value,columns=tt_cat_name)
```

```python
t_num_col=x_train.select_dtypes(include=np.number).columns
x_train_num=x_train[t_num_col].reset_index(drop=True)
tt_num_col=x_test.select_dtypes(include=np.number).columns
x_test_num=x_test[tt_num_col].reset_index(drop=True)
```

```python
x_train_final=pd.concat([x_train_num,t_cat_df.reset_index(drop=True)],axis=1)
x_test_final=pd.concat([x_test_num,tt_cat_df.reset_index(drop=True)],axis=1)
```

```python
model.fit(x_train_final,y_train)
y_train_pred=model.predict(x_train_final)
y_test_pred=model.predict(x_test_final)
test_acc=accuracy_score(y_test,y_test_pred)
train_acc=accuracy_score(y_train,y_train_pred)
print(train_acc)
print(test_acc)
```

```python
print('test')
print(confusion_matrix(y_test,y_test_pred))
print('precision_score:',precision_score(y_test,y_test_pred))
```

```python
print('train')
print(confusion_matrix(y_train,y_train_pred))
print('precision_score:',precision_score(y_train,y_train_pred))
```

```python
print(recall_score(y_test,y_test_pred))
```

```python
print(recall_score(y_train,y_train_pred))
```
