```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures,OneHotEncoder
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('student_performance_interactions (1).csv')
```

```python
x=df.drop(["student_id","final_score"],axis=1)
y=df["final_score"]
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
encode=OneHotEncoder()
```

```python
num_col=x.select_dtypes(exclude='object').columns
cat_col=x.select_dtypes(include='object').columns
```

```python
x_train[num_col]=scaler.fit_transform(x_train[num_col])
x_test[num_col]=scaler.fit_transform(x_test[num_col])
```

```python
t_label=encode.fit_transform(x_train[cat_col])
t_value=t_label.toarray()
```

```python
tt_label=encode.fit_transform(x_test[cat_col])
tt_value=tt_label.toarray()
```

```python
name=encode.get_feature_names_out()
```

```python
t_cat_df=pd.DataFrame(t_value,columns=name)
tt_cat_df=pd.DataFrame(tt_value,columns=name)
```

```python
x_train_final=pd.concat([x_train[num_col].reset_index(drop=True),t_cat_df.reset_index(drop=True)],axis=1)
x_test_final=pd.concat([x_test[num_col].reset_index(drop=True),tt_cat_df.reset_index(drop=True)],axis=1)
```

```python
model=LinearRegression()
model.fit(x_train_final,y_train)
```

```python
y_train_pred=model.predict(x_train_final)
y_test_pred = model.predict(x_test_final)
print( mean_squared_error(y_test, y_test_pred))
print( r2_score(y_test, y_test_pred))
```
